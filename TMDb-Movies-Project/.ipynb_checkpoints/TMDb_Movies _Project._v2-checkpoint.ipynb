## Data Cleaning
### Steps:
*    **Dropping the unwanted columns.**
*    **Checking for messing values.**
*    **Replacing or droping the NaN values.**
*    **Checking for duplicates.**
*   **Removing duplicates.**
*   **Adding Profit column.**
*   **seprating the pipe("|") values to new columns.**
    

```python
#Getting the columns names into a list:
Tmdb_df.columns.tolist()
```

**Dropping** the unwanted columns :

```python
#creating a list that ha the coulmns name that will be dropped
droped_col=[ 'id', 'imdb_id', 'budget_adj', 'revenue_adj', 'homepage', 'keywords', 'overview', 'production_companies', 'vote_count','tagline']
Tmdb_df=Tmdb_df.drop(droped_col,True)
#checking if the changes were made
Tmdb_df.head(2)
```

**Getting** the number of the NAN values

```python
Tmdb_df.isnull().sum()
```

```python
#dropping the NAN values
Tmdb_df=Tmdb_df.dropna()
```

```python
##checking if the changes were made
Tmdb_df.isnull().sum()
```

### Creating a histogram for each column to get a second look on the data:

```python
Tmdb_df.hist(figsize=(14,10));
```

```python
#Creating a new column "profit" which equal ('revenue'-'budget')
profit=Tmdb_df["revenue"]-Tmdb_df['budget']
#inserting the coulmn into the data frame
Tmdb_df.insert(loc=2,column="profit",value=profit)
Tmdb_df.head()
```

```python
#Checking for duplicates
Tmdb_df.duplicated().sum()
```

```python
#Droping duplicated entries
Tmdb_df.drop_duplicates(inplace=True)
#Checking if the changes were made
Tmdb_df.duplicated().sum()
```

```python
Tmdb_df.shape
```

* **Removing** films that have a run time equal to zero, more than 6 hours, or less than 10 minutes.

```python
# Getting the new data frame with the wanted filter applied to the original data frame
drop_df=Tmdb_df.query('(runtime==0) |(runtime>400)|(runtime<10)')
#droping the filtered data frame from the main data frame
Tmdb_df.drop(labels=drop_df.index,inplace=True)
#Resiting the indexes because it is missed up after removing the rows above
Tmdb_df.reset_index(inplace=True)
#droping the old index column
Tmdb_df.drop(columns='index',inplace=True)
Tmdb_df.shape
```

* **Removing** the films with messing revenue and budget columns.

```python
# Getting the new data frame with the wanted filter applied on the original data frame
drop2_df=Tmdb_df.query('(budget==0) | (revenue==0)')
#droping the filtered data frame from the main data frame
Tmdb_df.drop(labels=drop2_df.index,inplace=True)
#Resiting the indexes because it is missed up after removing the rows above
Tmdb_df.reset_index(inplace=True)
#droping the old index column
Tmdb_df.drop(columns='index',inplace=True)
Tmdb_df.shape
```

## Exploratory Data Analysis
## Research Question1: What is happening to the movie industry and people's ratings through the years?

```python
# We get the avg of profit and people ratings grouped by years of release
avg_profit_year=Tmdb_df.groupby('release_year')['profit'].mean()
avg_rating_year=Tmdb_df.groupby('release_year')['vote_average'].mean()
# Visulizing the data
#creating a figure using plt
fig = plt.figure(figsize=(15,8));
#making subplot 1
ax1 = fig.add_subplot(121);
#making subplot 1
ax2 = fig.add_subplot(122);
#ploting the wanted data and adjusting figure line color
ax1.plot(avg_profit_year,color="red");
ax2.plot(avg_rating_year);
#setting a title name and labels for X and Y axis for the first figure
ax1.set_title('Average Profits VS Years');
ax1.set_xlabel('Years');
ax1.set_ylabel('Average Profits');
#setting a title name and labels for X and Y axis for the second figure
ax2.set_title('Average Ratings VS Years');
ax2.set_xlabel('Years');
ax2.set_ylabel('Average Ratings');
```

## Conclusion
From the **(Average Profits VS Years)** figure we saw that the profit is increasing which means the film industry is making more money as time goes by, But people are becoming more selective in their ratings **(which is decreased as time goes by)** and the artistic sense became more distinguished.

## Research Questions 2&3:which genres are the most popular genres and which of them makes more profit?

```python
genres_df = Tmdb_df.assign(genres=Tmdb_df['genres'].str.split('|')).explode('genres')
genres_df.head()
```

```python
# visulizing what is the most repeated genre in the film industry
genres_df['genres'].value_counts().plot(kind='bar',figsize=(15,8));
```

```python
#creating a data frame with genres as indexes and the average profit of that genre as a value
profit_mean=genres_df.groupby('genres').profit.mean()
#creating a new bar chart
fig = plt.figure(figsize=(22,10))
plt.bar(profit_mean.index,profit_mean.values,align='center',color='teal');
#setting a title name and labels for X and Y axis for the figure
plt.ylabel('Avg Profit');
plt.xlabel('Genres');
plt.title('Avg Profit vs Genres');
```

```python
#creating a data frame with genres as indexes and the average ratings of that genre as a value
ratings_mean=genres_df.groupby('genres').vote_average.mean()
#creating a new bar chart
fig = plt.figure(figsize=(25,10))
plt.bar(ratings_mean.index,ratings_mean.values,color='darkkhaki');
#setting a title name and labels for X and Y axis for the figure
plt.ylabel('Avg Ratings');
plt.xlabel('Genres');
plt.title('Avg Ratings vs Genres');
```

```python
#creating a data frame with genres as indexes and the average popularity of that genre as a value
pop_mean=genres_df.groupby('genres').popularity.mean()
#creating a new bar chart
fig = plt.figure(figsize=(30,15))
plt.bar(pop_mean.index,pop_mean.values,color='navy');
#setting a title name and labels for X and Y axis for the figure
plt.ylabel('Avg Popularity',fontsize=20);
plt.xlabel('Genres',fontsize=20);
plt.title('Avg Popularity vs Genres',fontsize=20);
```

### Conclusion:
**Animation** and **Adventure** films often have the most profit but the heights average ratings go for **Documentary** films, **Science fiction** and **Adventure** films have the most popularity, most films have **Drama** genre in it (over 4000 films).

## Research Question4:What is the average run time of the films and which of them is more popular?

## Research Question5:What  are the top 10 movies with the highest profit and budget?

```python
#top 10 films which made highest Profit.

#makeing a new data frame with only the profit coulmn values sorted
info = pd.DataFrame(Tmdb_df['profit'].sort_values(ascending = False))

#setting both data frames title to be equal to each other
info['original_title'] = Tmdb_df['original_title']
data = list(map(str,(info['original_title'])))

#extract the top 10 movies with high revenue data from the list and dataframe.
x = list(data[:10])
y = list(info['profit'][:10])

#make the point plot and setup the title and labels.
ax= sns.barplot(x=y,y=x)
sns.set(rc={'figure.figsize':(12,5)})
ax.set_title("Top 10 Highest Profit Movies",fontsize = 20)
ax.set_xlabel("profit",fontsize = 13)
ax.set_ylabel("Movie Name",fontsize = 13)
sns.set_style("darkgrid")
```

```python
#top 10 films which made highest budget.
#makeing a new data frame with only the budget coulmn values sorted
info = pd.DataFrame(Tmdb_df['budget'].sort_values(ascending = False))
info['original_title'] = Tmdb_df['original_title']
data = list(map(str,(info['original_title'])))

#extract the top 10 movies with high revenue data from the list and dataframe.
x = list(data[:10])
y = list(info['budget'][:10])

#make the point plot and setup the title and labels.
ax= sns.barplot(x=y,y=x)
sns.set(rc={'figure.figsize':(12,5)})
ax.set_title("Top 10 High budget Movies",fontsize = 20)
ax.set_xlabel("budget",fontsize = 13)
ax.set_ylabel("Movie name",fontsize = 13)
sns.set_style("darkgrid")
```

### Conclusion
* **Spending more money on the film budget doesn't mean that the movie will achieve a high profit.**
